**Choose one continuous variable and one categorical variable, and plot them six different ways.**

```python
sns.set(style="whitegrid")

g = sns.catplot(x="country", y="ad_views", hue="country", data=df,
                   height=5, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("number of views")
g.set_xlabels("nation")
plt.title('Pointplot: Ad View by Nation')
plt.show()
```

```python
a= sns.catplot(x="country", y="ad_views", hue="country", data=df, height=6, kind="bar", palette="pastel", ci = 95)
a.despine(left=True)
a.set_ylabels("number of views")
a.set_xlabels("nation")
plt.title('Barplot: Views by Nation')
plt.show()

#ci = 95 means that there is a 95% confidence that US has 15-58 views? 
#height of the bars shows avg ad view
```

```python
f = sns.stripplot(x="country", y="ad_views", data=df)
plt.title('Stripplot: Views by Nation')
```

```python
g = sns.boxplot(x="country", y="ad_views", data=df)
plt.title('Boxplot: Views by Nation')
# the box shows the interquartile range (1stQ to 3rdQ). The line is median number
#pros:compare different categories to get general overview of median 
```
